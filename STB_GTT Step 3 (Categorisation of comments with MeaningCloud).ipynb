```python
import pandas as pd
import numpy as np
import string

from pprint import pprint
```

```python
comments_master = pd.read_csv("/Users/chrtjen/Desktop/STB_GTT/Step3/stb_gtt_0-1000.csv", engine='python')

comments_master
```

```python
to_categorise = comments_master['Discussion_translated']
```

```python
#run the API
#initialise empty list to store the results of the run
#iterate through the column, appending results from each row into the list

import requests

url = "https://api.meaningcloud.com/deepcategorization-1.0"

results = []
for i in to_categorise:
    text = str(i)
    payload={
        'key': 'addc68868e4b5945047ff2ad0c10f38a',
        'txt': text,
        'model': 'IAB_2.0-tier4',  # like IAB_2.0_en
        }

    response = requests.post(url, data=payload)
    print('Status code:', response.status_code)
    print(response.json())
    
    results.append(response.json().get('category_list'[].label))
```

```python
#insert list into main dataframe
comments_master['Categorisation'] = results

comments_master.to_csv("/Users/chrtjen/Desktop/STB_GTT/Categorised Comments_0-1000.csv")
```

```python
combined = {}

for i in range (len(comments_master['Categorisation'])):
    list_of_ds = comments_master['Categorisation'][i]
    for b in list_of_ds:
        for key,val in b.items():
            if key not in combined.keys():
                combined[key] = val
            else: combined[key].extend(val)
```

```python
new={}
for a in comments_master['Categorisation']:
    for b in a:
        for i,j in b.items():
            new.setdefault(i,[]).extend(j)
```
