```python
import pandas as pd
import numpy as np
import string

from pprint import pprint
```

```python
comments_master = pd.read_csv("/Users/chrtjen/Desktop/STB_GTT/stb_gtt_sample2.csv", engine='python')

comments_master
```

```python
#convert cells that became empty into NaN
comments_master['Discussion_cleaned'].replace('',np.NaN,inplace=True)

#remove null rows
comments_master.dropna(subset=['Discussion_cleaned'],inplace=True)

#reset index
comments_master = comments_master.reset_index(drop=True)

comments_master
```

```python
listofcomments = comments_master['Discussion_translated']

listofcomments
```

```python
import requests

url = "https://api.meaningcloud.com/deepcategorization-1.0"

status = []
for i in listofcomments:
    text = str(i)
    payload={
        'key': 'addc68868e4b5945047ff2ad0c10f38a',
        'txt': text,
        'model': 'IAB_2.0-tier4',  # like IAB_2.0_en
        }

    response = requests.post(url, data=payload)
    print('Status code:', response.status_code)
    print(response.json())
    
    status.append(response.json())
```
