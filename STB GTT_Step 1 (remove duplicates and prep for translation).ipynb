```python
path ='/Users/chrtjen/Users/chrtjen/Desktop/STB_GTT/STB_GTT_Step1Output'
filenames = os.listdir()

dfs = []
df_names = []
for filename in filenames:
    if 'csv' in filename:
        df_names.append(filename)
        dfs.append(pd.read_csv(filename))
  
print(f'read the following dataframes: {df_names}')
```

```python
#create the master dataframe

comments_master = pd.concat(dfs, ignore_index=True, sort=False).dropna(subset=['Discussion_translated']).reset_index(drop=True)
comments_master
```

```python
#get duplicated comments

duplicates = comments_master[comments_master.duplicated()]
duplicates_index = list(duplicates.index.values)
duplicates
```

#### Check index numbers in duplicate list, open csv with the corresponding title of the comments in that index number range and manually delete it


#### Important to manually remove duplicates as they will lenghten time needed for translation. If translation is not needed, no issue to just use .drop to remove them from the dataset.

```python
#this is just to check that only unique comments remain

comments_master = comments_master.drop(duplicates_index).reset_index(drop=True)
comments_master
```
