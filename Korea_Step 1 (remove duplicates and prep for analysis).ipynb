```python
import os
import pandas as pd
import numpy as np
import string

from pprint import pprint

os.chdir("/Users/chrtjen/Desktop/STB_GTT/Korea")
```

```python
filenames = os.listdir()

dfs = []
df_names = []
for filename in filenames:
    if 'csv' in filename:
        df_names.append(filename)
        dfs.append(pd.read_csv(filename))
  
print(f'read the following dataframes: {df_names}')
```

```python
#create the master dataframe

comments_master = pd.concat(dfs, ignore_index=True, sort=False).dropna(subset=['Discussion']).reset_index(drop=True)
comments_master
```

```python
#get duplicated comments

duplicates = comments_master[comments_master.duplicated()]
duplicates_index = list(duplicates.index.values)
duplicates
```
